```python
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
jsons = []
```

```python
for file in os.listdir("data/raw/"):
    data = json.load(open("data/raw/"+file))
    jsons.append(data)
```

```python
atribute = 'incendiaries'
```

```python
struct = pd.DataFrame.from_records(jsons[0][atribute])
struct_map_name = jsons[0]['map_name']
struct['map_name'] = struct_map_name

for i in range(len(jsons[0:])):
    to_concat = pd.DataFrame.from_records(jsons[i][atribute])
    to_concat_map_name = jsons[i]['map_name']
    to_concat['map_name'] = to_concat_map_name
    
    struct = pd.concat([struct, to_concat])
```

```python
for column in struct:
    if (column == "heatmap_point"):
        new_columns = pd.json_normalize(struct[column])
        struct.drop(columns=column, inplace=True)
        struct[new_columns.columns] = new_columns
```
