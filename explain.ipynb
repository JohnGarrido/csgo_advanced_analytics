```python
import os 
import pandas as pd
```

```python
parquets = os.listdir('data/matches/bronze_parquets')
```

```python
df = pd.read_parquet("data/matches/bronze_parquets/"+parquets[0]) 

for parquet in parquets:
    if(parquet != parquet[0]):
        newFrame = pd.read_parquet("data/matches/bronze_parquets/"+parquet) 
        df = pd.concat([df, newFrame])
```
